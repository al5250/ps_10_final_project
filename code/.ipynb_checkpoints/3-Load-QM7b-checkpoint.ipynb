```python
import numpy as np
import pandas as pd
from hf import compute_all_integrals, hartree_fock
from utils import load_basis_set
import time as time
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pickle as pickle

def write_pickle(obj, filename):
    pickle.dump(obj, open(filename, 'wb'))
```

```python
data_dict = {}
with open('data/qm7b/dsgdb7njp.xyz', 'r') as fp:
    line = fp.readline()
    counter = 0
    while line:
        n_nuclei = int(line)
        properties = fp.readline().split()
        properties = [float(x) for x in properties]
        atoms = []
        Rs = []
        for _ in range(n_nuclei):
            line = fp.readline().split()
            atoms.append(line[0])
            Rs.append([float(x) for x in line[1:]])
        data_dict[counter] = {
            'n_nuclei': n_nuclei,
            'atoms': np.array(atoms),
            'Rs': np.array(Rs),
            'ae_pbe0': properties[0],
            'p_pbe0': properties[1],
            'p_scs': properties[2],
            'homo_gw': properties[3],
            'homo_pbe0': properties[4],
            'homo_zindo': properties[5],
            'lumo_gw': properties[6],
            'lumo_pbe0': properties[7],
            'lumo_zindo': properties[8],
            'ip_zindo': properties[9],
            'ea_zindo': properties[10],
            'e1_zindo': properties[11],
            'emax_zindo': properties[12],
            'imax_zindo': properties[13]
        }
        np.array(Rs)
        np.array(atoms)
        
        counter += 1
        fp.readline()
        line = fp.readline()
df = pd.DataFrame(data_dict).T
```

```python
df['n_non_H'] = df['atoms'].apply(lambda x: np.sum(x != 'H'))
```

```python
atom2charge = {
    'H': 1,
    'He': 2,
    'Li': 3,
    'Be': 4,
    'B': 5,
    'C': 6,
    'N': 7,
    'O': 8,
    'F': 9,
    'Ne': 10,
    'Na': 11,
    'Mg': 12,
    'Al': 13,
    'Si': 14,
    'P': 15,
    'S': 16,
    'Cl': 17,
    'Ar': 18
}
```

```python
df['Zs'] = df['atoms'].apply(lambda x: np.array([atom2charge[a] for a in x]))
```

```python
df['n_electrons'] = df['Zs'].apply(lambda x: x.sum())
```

```python
basis_set = load_basis_set('basis_sets/STO-3G.txt')
```

```python
np.sum(df['n_non_H'] <= 4)
```

```python
out_dict = {}

for i in range(43, 59):
    row = df.iloc[i]
    Rs, Zs, atoms, n_nuclei, n_electrons = row['Rs'], row['Zs'], row['atoms'], row['n_nuclei'], row['n_electrons']

    ao_types = [basis_set[a]['orbital_types'] for a in atoms]
    ao_exps = [basis_set[a]['expos'] for a in atoms]
    ao_coefs = [basis_set[a]['coefs'] for a in atoms]

    orbital_counts = [len(at) for at in ao_types]
    ao_types = np.hstack(ao_types)
    ao_exps = np.vstack(ao_exps)
    ao_coefs = np.vstack(ao_coefs)
    ao2nuc = np.repeat(np.arange(n_nuclei), orbital_counts, axis=0)
    n_orbitals = ao2nuc.shape[0]

    t = time.time()
    out = hartree_fock(n_electrons, ao_exps, ao_coefs, ao_types, Rs, Zs, ao2nuc)
    delta_t = time.time() - t
    
    out_dict[i] = (out, delta_t)
#     write_pickle(out_dict, 'hf_output2.p')
    print(i)
```

```python
atomic_energy_dict = {}
atomic_energy_list = np.zeros((18,))

for a in atom2charge.keys():
    Rs = np.array([
            [0, 0, 0]
        ])
    Z = atom2charge[a]
    Zs = np.array([[Z]])
    atoms = np.array([a])
    n_nuclei = 1
    if Z % 2 == 0:
        n_electrons = Z
    elif Z % 2 == 1:
        n_electrons = Z + 1

    ao_types = [basis_set[a]['orbital_types'] for a in atoms]
    ao_exps = [basis_set[a]['expos'] for a in atoms]
    ao_coefs = [basis_set[a]['coefs'] for a in atoms]

    orbital_counts = [len(at) for at in ao_types]
    ao_types = np.hstack(ao_types)
    ao_exps = np.vstack(ao_exps)
    ao_coefs = np.vstack(ao_coefs)
    ao2nuc = np.repeat(np.arange(n_nuclei), orbital_counts, axis=0)
    n_orbitals = ao2nuc.shape[0]

    out = hartree_fock(n_electrons, ao_exps, ao_coefs, ao_types, Rs, Zs, ao2nuc)
    
    print(a)
    
    if Z % 2 == 0:
        atomic_energy_dict[a] = out[0]
        atomic_energy_list[Z-1] = out[0]
    else:
        atomic_energy_dict[a] = out[0] - out[1][n_electrons // 2 - 1]
        atomic_energy_list[Z-1] = out[0] - out[1][n_electrons // 2 - 1]
        
```

```python
plt.figure(figsize=(8, 6))
x = df_small['n_orbitals'].astype('float')
y = df_small['comp_time']
plt.plot(x, y, 'o')
x_grid = np.linspace(0, 35, 1000)
strs = []
for deg in range(1, 5):
    M = np.column_stack(([1] * len(x), x ** deg))
    k, _, _, _ = np.linalg.lstsq(M, y, rcond=None)
    if k[0] >= 0:
        k_0 = '+' + str(k[0].round(2))
    else:
        k_0 = str(k[0].round(2))
    strs.append('$y = {} \cdot x^{} {}$'.format(k[1].round(5), deg, k_0))
    plt.plot(x_grid, k[0] + k[1] * x_grid ** deg)
plt.ylim([-200, 1400])
plt.xlim([5, 35])
plt.legend(['observed values'] + strs)
plt.xlabel('Number of Orbitals')
plt.ylabel('Computation Time (s)')
```
